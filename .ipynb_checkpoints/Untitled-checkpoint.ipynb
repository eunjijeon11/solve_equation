# **ÏõêÎ¶¨**üìö

Í≥°ÏÑ† y = f(x) ÏúÑÏùò Ï†ÅÎãπÌïú Ï†ê (x1,f(x1))ÏóêÏÑúÏùò Ï†ëÏÑ†Ïùò Î∞©Ï†ïÏãùÏùÄ  
y = f'(x1)(x - x1) + f(x1) ÏûÖÎãàÎã§.  
y = f'(x1)x - f'(x1)x1 + f(x1) Ïù¥ÎØÄÎ°ú yÏûêÎ¶¨Ïóê 0ÏùÑ ÎÑ£Í≥† Ï†ïÎ¶¨ÌïòÎ©¥  
x = x1 - f(x1)/f'(x1) ÏûÖÎãàÎã§. Ïó¨Í∏∞ÏóêÏÑú xÎäî xÏ†àÌé∏Ïù¥ Îê©ÎãàÎã§.  
ÏúÑ xÎ•º x2Î°ú ÏÑ§Ï†ïÌïòÍ≥† Ïù¥ Í≥ºÏ†ïÏùÑ Î∞òÎ≥µÌïòÎ©¥ xnÏù¥ Ï†êÏ†ê Î∞©Ï†ïÏãùÏùò Ìï¥Ïóê Í∞ÄÍπåÏõåÏßÄÍ≤å Îê©ÎãàÎã§.

Í∑∏Îü∞Îç∞ ÎßåÏïΩ f'(xn)Ïù¥ 0Ïù¥ ÎÇòÏò®Îã§Î©¥ ÏãùÏùÑ Í≥ÑÏÇ∞Ìï† Ïàò ÏóÜÏùÑ Í≤ÉÏûÖÎãàÎã§.  
Ïù¥Î•º Î∞©ÏßÄÌïòÍ∏∞ ÏúÑÌï¥ Í∏∞Ïö∏Í∏∞Í∞Ä 0Ïù¥ ÎÇòÏò§Î©¥ - f(xn-1)/f'(xn-1)ÎßåÌÅº Îã§Ïãú Ïù¥ÎèôÏãúÏºúÏÑú Í≥ÑÏÇ∞ÏùÑ ÌïòÎèÑÎ°ù ÌïòÏòÄÏäµÎãàÎã§.  
(xn-1ÏùÄ Ïù¥Ï†ÑÏùò xÏùò ÏúÑÏπò)  
ÎßåÏïΩ f'(x1)Ïù¥ 0Ïù¥ÎùºÎ©¥ x1Í∞íÏóê 1ÏùÑ ÎçîÌï¥Ï£ºÏóàÏäµÎãàÎã§.

```python
from sympy import *
%matplotlib inline

init_printing()
```

```python
def solve_equation(f, epoch = 10, t = 0, graph = False, xlim = (0,0), ylim = (0,0)) :
    x = Symbol('x')
    f_diff = diff(f, x)
    move = 0
    
    for i in range(epoch) :
        if (f_diff.subs(x,t) == 0) :
            if (move == 0) :
                move += 1
            t += move
            continue
        
        if (graph) :
            line = expand(f_diff.subs(x,t)*(x-t) + f.subs(x,t))
            plot(f,line, xlim=xlim, ylim=ylim)
        
        move = - f.subs(x,t)/f_diff.subs(x,t)
        t += move
        
        if (f.subs(x,t) == 0) :
            break
        print(N(t))
        
    print("Í∑ºÏùò Í∑ºÏÇ¨Í∞íÏùÄ", N(t))
    real = [N(m) for m in solve(f)]
    print("Ïã§Ï†ú Í∑ºÏùÄ", real)
```

# **Ïã§Ïäµ**üê•
Ïù¥Ï†úÎäî Ïó¨Îü¨Í∞ÄÏßÄ Ìï®ÏàòÏùò Í∑ºÏùò Í∑ºÏÇ¨Í∞íÏùÑ Íµ¨Ìï¥Î≥¥Í≥† ÌååÏù¥Ïç¨ ÎÇ¥Ïû•Ìï®ÏàòÎ•º Ïù¥Ïö©Ìï¥ÏÑú Íµ¨ÌïòÎäî Ïã§Ï†ú Í∑ºÍ≥º ÎπÑÍµêÌï¥Î≥º Í≤ÉÏûÖÎãàÎã§.  
Ïã§Ï†ú Í∑ºÏóêÎäî Î≥µÏÜåÏàòÎèÑ Ìè¨Ìï®ÎêòÏßÄÎßå Í∑ºÏÇ¨Í∞íÏùÄ Ïã§ÏàòÎßå ÎÇòÏò§ÎãàÍπå Ï∞∏Í≥†Ìï¥Ï£ºÏÑ∏Ïöî!

### **Îã§Ìï≠Ìï®Ïàò Í∑º Íµ¨ÌïòÍ∏∞**  
2x^2 + x -1 = 0

```python
solve_equation(2*x**2 + x - 1)
```

ÏúÑ Î∞©Ï†ïÏãùÏùÄ Ïã§Ï†úÎ°úÎäî Í∑ºÏù¥ Îëê Í∞úÏù∏Îç∞, ÍµêÍ≥ºÏÑú Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Ï≤òÏùå ÏãúÏûë ÏúÑÏπòÏóê Îî∞Îùº ÌïòÎÇòÏùò Í∑ºÎ∞ñÏóê Íµ¨Ìï† Ïàò ÏóÜÏäµÎãàÎã§.  
Í∑ºÏùÑ Î™®Îëê Íµ¨ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî Ï¥àÍ∏∞ ÏÑ§Ï†ïÍ∞íÏùÑ Îã§ÏñëÌïòÍ≤å Ï§òÏÑú Ï†ïÌôïÎèÑÎ•º ÎÜíÏù¥Îäî Î∞©Î≤ïÏùÑ ÏÇ¨Ïö©Ìï† Ïàò ÏûàÏäµÎãàÎã§.

#### Í∑∏ÎûòÌîÑÎ°ú ÏïåÏïÑÎ≥¥Í∏∞üìà

```python
solve_equation(2*x**2 + x - 1, graph=True, xlim=(-2,2), ylim=(-3,3))
```
